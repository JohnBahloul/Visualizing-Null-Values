# Visualizing Null Values

```python
%matplotlib inline 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
mental_health_data=pd.read_csv('survey.csv')
mental_health_data.head()
```

```python
# find number of nan values per column
mental_health_data.isnull().sum()
```

```python
msno.matrix(mental_health_data)
plt.show()
```

```python
msno.bar(mental_health_data)
plt.show()
```

```python
# bin variables against eachother to view the nullity correlation
msno.dendrogram(mental_health_data)
plt.show()
```

## Remove columns

```python
clean_data = mental_health_data.drop(columns = ['state','work_interfere','comments'])
clean_data.head()
```

```python
# notice after removing the columns we now have no null values
msno.matrix(clean_data)
plt.show()
```

```python
clean_data.keys()
```

```python
clean_data.info()
```

## Let us look at the Age Distribution

```python
age = clean_data['Age']
age.value_counts().shape
```

```python
clean_age = []

for i in range(0, len(age)):
    if age.at[i] > 11 and age.at[i] < 329:
        clean_age.append(age.at[i])
```

```python
new_clean_age = pd.Series(clean_age)
new_clean_age.value_counts().shape
```

```python
age_hist = new_clean_age.plot.hist(color='blue')
age_hist.set_title('Age Distribution', fontsize=14)
age_hist.set_xlabel('Age')
plt.show()
```

```python
# Alternatively, use a rug and kde plot to view the distribution and probaility density.
sns.kdeplot(new_clean_age, shade=True);
sns.rugplot(new_clean_age);
```
